```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv('products.csv')
```

```python
df
```

```python
df.isnull().sum()
```

### Remove unncessary columns

```python
del df['product_link']
del df['img_link']
del df['meta_description']
del df['meta_keyword']
```

```python
df
```

### Average rating of a product(non zero)

```python
rat=0
count=0
for i in df['rating']:
    if i!=0.0:
        rat+=i
        count+=1
print(round(int(rat)/count,2))
        
```

### Most expensive product(Marked price) 

```python
df[df['marked_price']==df['marked_price'].max()]
```

### Most expensive product(discounted price)

```python
df[df['discounted_price']==df['discounted_price'].max()]
```

### Most discounted product in terms of percentage

```python
df[df['discount_percent']==df['discount_percent'].max()]
```

### Most discounted amount in terms of discounted numbers

```python
df[df['discount_amount']==df['discount_amount'].max()]
```

### Sort rows by Discount percentage 

```python
df.sort_values(by='discount_percent',ascending=False)
```

### Finding all the brands

```python
len(df['brand_name'].unique())
```

### Finding the brand with most product numbers

```python
data=[]
for brand_name in df['brand_name'].unique():
  c=0
  for i in df['brand_name']:
      if i==brand_name:
          c+=1
  data.append([brand_name,c])       
data=pd.DataFrame(data,columns=['name','product'])
data
```

```python
data.sort_values(by='product',ascending=False).head(10)
```

### Finding the brand with most product numbers using groupby

```python
df.groupby('brand_name').size().sort_values(ascending=False).head(10)
```

### Finding the most expensive brand

```python
df.groupby('brand_name')['marked_price'].mean().round().sort_values(ascending=False).head(10)
```

### Brands gives most discount

```python
df.groupby('brand_name')['discounted_price'].mean().round().sort_values(ascending=False).head(10)
```

```python
df.groupby('brand_name')['discount_percent'].mean().round().sort_values(ascending=False).head(10)
```

```python
df_pr=df[df['rating']!=0]
df_pr.groupby('brand_name')['rating'].mean().sort_values(ascending=False).head(10)
```

### Number of rating

```python
df_pr=df[df['rating_count']!=0]
df_pr.groupby('brand_name')['rating_count'].mean().round().sort_values(ascending=False).head(10)
```

```python
for i in df['seller_address']:
    print(i.split(' '))
```

```python
pincode=[]
state=[]
for i in df['seller_address']:
    if i!='Not Mentioned':
        pincode.append(i.split(' ')[-1])
        state.append(i.split(',')[-1].split('-')[0].strip())
    else:
        pincode.append(np.nan)
        state.append(np.nan)
df['pincode']=pincode
df['state']=state
```
